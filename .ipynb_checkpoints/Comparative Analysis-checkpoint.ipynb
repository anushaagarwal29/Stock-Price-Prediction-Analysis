# Comparative Analysis of Predictive Models

Here, we present a comparative analysis of several models trained to predict stock prices. The models considered in this analysis are ARIMA, LSTM, LSTM + Sentiment, Transformer (with only Adjusted Close Price), and Transformer (with 7 features). The key performance metrics used for comparison are Mean Squared Error (MSE), Mean Absolute Percentage Error (MAPE), and R² Score.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Example: Dictionary of metrics for each model
# metrics_data = {
#     'Model': ['ARIMA', 'LSTM', 'LSTM + Sentiment', 'Transformer (Adj Close)', 'Transformer (7 features)'],
#     'Training MSE': [31.516749, 0.170396, 0.154271, 0.118643, 0.116749],
#     'Validation MSE': ['-', 0.682905, 0.622456, 0.565707, 0.656560],
#     'Test MSE': [178.536682, 1.456814, 1.597258, 0.511887, 0.508978],
#     'Training MAPE': ['-', 0.009484, 0.008848, 0.007439, 0.007845],
#     'Validation MAPE': ['-', 0.013258, 0.012473, 0.011258, 0.012908],
#     'Test MAPE': [0.202662, 0.016370, 0.016400, 0.010247, 0.010358],
#     'Training R²': [0.586365, 0.996362, 0.996706, 0.997461, 0.997542],
#     'Validation R²': ['-', 0.981402, 0.983048, 0.984602, 0.982176],
#     'Test R²': [-1.398789, 0.975440, 0.973073, 0.991343, 0.991449]
# }

metrics_data = {
    'Model': ['ARIMA', 'LSTM', 'LSTM + Sentiment', 'Transformer (Adj Close)', 'Transformer (7 features)', 'GPT (Adj Close)'],
    'Training MSE': [31.516749, 0.170396, 0.154271, 0.118643, 0.116749, '-'],
    'Validation MSE': ['-', 0.682905, 0.622456, 0.565707, 0.656560, '-'],
    'Test MSE': [178.536682, 1.456814, 1.597258, 0.511887, 0.508978, 0.708104],
    'Training MAPE': ['-', 0.009484, 0.008848, 0.007439, 0.007845, '-'],
    'Validation MAPE': ['-', 0.013258, 0.012473, 0.011258, 0.012908, '-'],
    'Test MAPE': [0.202662, 0.016370, 0.016400, 0.010247, 0.010358, 0.012459],
    'Training R²': [0.586365, 0.996362, 0.996706, 0.997461, 0.997542, '-'],
    'Validation R²': ['-', 0.981402, 0.983048, 0.984602, 0.982176, '-'],
    'Test R²': [-1.398789, 0.975440, 0.973073, 0.991343, 0.991449, 0.987729]
}

# Convert to DataFrame
metrics_df = pd.DataFrame(metrics_data)
```

## Model Performance Metrics

The following table presents the performance metrics for each model across the training, validation, and test datasets.


```python
# Replace '-' with NaN so that it doesn't plot for those values
metrics_df = metrics_df.replace('-', np.nan)

# Display the DataFrame for performance comparison
metrics_df.set_index('Model', inplace=True)
metrics_df
```

```python
import numpy as np

# Function to add labels on top of the bars
def add_value_labels(ax, spacing=5, rotation=0):
    """Add labels to the bars."""
    for rect in ax.patches:
        y_value = rect.get_height()
        if np.isnan(y_value) or y_value==0:  # Skip if the value is NaN
            continue
        x_value = rect.get_x() + rect.get_width() / 2
        space = spacing if y_value > 0 else -spacing
        label = "{:.2f}".format(y_value)

        # Adjust the display to avoid overlapping
        ax.annotate(
            label,
            (x_value, y_value),
            xytext=(0, space),
            textcoords="offset points",
            ha="center",
            va="bottom",
            fontsize=8,  # Adjust fontsize
            rotation=rotation  # Rotate the labels if needed
        )

# Plotting all metrics in a row
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Adjust figure size for better spacing

# MSE Plot
ax1 = metrics_df[['Training MSE', 'Validation MSE', 'Test MSE']].plot(kind='bar', ax=axes[0, 0], color=['#1f77b4', '#ff7f0e', '#2ca02c'])
ax1.set_title('Mean Squared Error (MSE) Comparison')
ax1.set_ylabel('MSE')
ax1.set_xticklabels(metrics_df.index, rotation=45, ha='right')
ax1.set_ylim(0, ax1.get_ylim()[1] * 1.25)  # Increase the upper limit by 10%
add_value_labels(ax1, spacing=5, rotation=45)  # Rotate labels

# MAPE Plot
ax2 = metrics_df[['Training MAPE', 'Validation MAPE', 'Test MAPE']].plot(kind='bar', ax=axes[0, 1], color=['#1f77b4', '#ff7f0e', '#2ca02c'])
ax2.set_title('Mean Absolute Percentage Error (MAPE) Comparison')
ax2.set_ylabel('MAPE')
ax2.set_xticklabels(metrics_df.index, rotation=45, ha='right')
ax2.set_ylim(0, ax2.get_ylim()[1] * 1.25)  # Increase the upper limit by 10%
add_value_labels(ax2, spacing=5, rotation=45)

# R² Plot
ax3 = metrics_df[['Training R²', 'Validation R²', 'Test R²']].plot(kind='bar', ax=axes[1, 0], color=['#1f77b4', '#ff7f0e', '#2ca02c'])
ax3.set_title('R² Score Comparison')
ax3.set_ylabel('R²')
ax3.set_xticklabels(metrics_df.index, rotation=45, ha='right')
ax3.set_ylim(ax3.get_ylim()[0], ax3.get_ylim()[1] * 1.4)  # Increase the upper limit by 5%
add_value_labels(ax3, spacing=5, rotation=45)

# Remove the unused axis (bottom right)
fig.delaxes(axes[1, 1])

# Adjust layout for spacing
plt.tight_layout()
plt.subplots_adjust(top=1.2, hspace=0.6)  # Increase space between subplots and top

plt.show()
```

### Model Performance Comparison:

1. **Mean Squared Error (MSE) Comparison**:
   - ARIMA shows the highest test MSE. This makes it the least effective model in terms of accuracy.
   - All LSTM-based and Transformer models show significantly lower MSE on both training and test data.
   - Transformer models, especially with multi-feature inputs, demonstrate superior performance with minimal MSE on the test set.

2. **Mean Absolute Percentage Error (MAPE) Comparison**:
   - The ARIMA model has a very high MAPE of 0.2, indicating poor prediction quality in percentage terms.
   - LSTM and Transformer models achieve near-zero MAPE, showing they effectively minimize error percentage across the board.

3. **R² Score Comparison**:
   - The ARIMA model's R² score is negative, indicating that it performs worse than a simple mean prediction.
   - The R² scores for all LSTM-based and Transformer models are near 1.0, indicating that these models explain nearly all the variance in both the training and test data.
   - Both LSTM and Transformer models deliver outstanding performance with no significant drop-off from training to test data.
