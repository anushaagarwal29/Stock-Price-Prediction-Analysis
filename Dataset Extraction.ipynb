```python
import pandas as pd
import numpy as np
```

# 1. Apple Twitter Data Extraction

```python
# Extract tweets from tweet.csv file into a dataframe
tweet_df = pd.read_csv('Tweet.csv')
print(len(tweet_df))
tweet_df.head()
```

```python
# Extract tweets about AAPL from twitter dataset
aapl_tweets = tweet_df[tweet_df['body'].str.contains('AAPL', case=False, na=False)]
aapl_tweets['post_date'] = pd.to_datetime(aapl_tweets['post_date'], unit='s')
aapl_tweets.head()
```

```python
# Save the AAPL tweets to csv
aapl_tweets.to_csv('AAPL_Tweets.csv', index=False)
```
